# Classification of app reviews using NLP

### 1) import data

```python
import numpy
import pandas as pd
```

__Dataset 1 (Maalej Dataset)__: The truth dataset contains 3691 reviews from different Google’s apps store and Apple’s app store. All reviews were classified into four classes related to Software engineering’s maintenance task as follows: 252 reviews as **Feature Request (FR)**, 370 reviews as **bug report/Problem Discovery (BR/PD)**, 607 reviews as **User Experience (UE)** and 2461 reviews as **Rating (RT)**.

```python
df1 = pd.read_excel("./data/Maalej Dataset.xlsx", index_col=0)
# df1 = df1[['review', 'stopwords_removal_lemmatization', 'task']]
df1 = df1[['review', 'task']]
```

```python
df1[df1['task'] == "PD"]
```

```python
df1.loc[df1['task'] == "RT", 'task'] = "rating"
df1.loc[df1['task'] == "FR", 'task'] = "feature request"
df1.loc[df1['task'] == "PD", 'task'] = "problem discovery"
df1.loc[df1['task'] == "UE", 'task'] = "user experience"
df1
```

```python
# rename task to class
df1 = df1.rename(columns={"task": "class"})
df1
```

__Dataset 2 (Pan Dataset)__: This dataset consist of with 1390 reviews from all previously mentioned apps and all reviews were classified into four classes related to Software engineering’s maintenance task as follows: 192 reviews as **Feature Request (FR)**, 494 reviews as **Problem Discovery (PD)**, 603 reviews as **Information Gaing (IG)** and 101 reviews as **Information Seeking (IS)**.

```python
df2 = pd.read_excel("./data/Pan Dataset.xlsx", index_col=0)
df2
```

## Combining two dataframes 

```python
df = pd.concat([df1, df2])
df.index.name = "id"
df
```
